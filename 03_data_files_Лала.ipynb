3. По данным из файла `addres-book-q.xml` сформировать список словарей с телефонами каждого из людей. 

````python

import re
#регулярное выражение для телефонных номеров
phone_regex = re.compile(r'(?<!\d)\d{3,}(?!\d)')


def extract_phones(xml_file):
    with open(xml_file, 'r') as f:
        data = f.read()
        #все совпадения с шаблоном телефонного номера в данных
        matches = phone_regex.findall(data)
        return matches

print(extract_phones('addres-book-q.xml'))

````

### pickle

2.1 На основе файла `contributors_sample.json` создайте словарь следующего вида: 
```
{
    должность: [список username людей, занимавших эту должность]
}
```

````python

# Создание словаря
positions = {}
for contributor in data:
    # Получение должности из словаря contributor
    position = contributor['jobs']
    username = contributor['username']
    # Преобразованиясписка в кортеж
    position_tuple = tuple(position)
    if position_tuple in positions:
        # Добавление имени пользователя в список, связанный с ключом position_tuple
        positions[position_tuple].append(username)
    else:
        positions[position_tuple] = [username]

# Вывод словаря
print(positions)
````

````python
import json
import pickle

# Преобразование кортежей в строки в позициях словаря
positions_str_keys = {str(k): v for k, v in positions.items()}

#  pickle
with open('job_people.pickle', 'wb') as f:
    pickle.dump(positions, f)

#  JSON 
with open('job_people.json', 'w') as f:
    json.dump(positions_str_keys, f, indent=4)
# Чтение данных из файла 
with open('job_people.json', 'r') as f:
    data1 = json.load(f)
    print(data1)


with open('job_people.pickle', 'rb') as f:
    data2 = pickle.load(f)
    print(data2)
````

### XML

3.1 По данным файла `steps_sample.xml` сформируйте словарь с шагами по каждому рецепту вида `{id_рецепта: ["шаг1", "шаг2"]}`. Сохраните этот словарь в файл `steps_sample.json`

````python

from nltk import flatten
from bs4 import BeautifulSoup
import json
# Загрузка XML-файла с рецептами
with open("steps_sample.xml") as f:
    ab = BeautifulSoup(f, "xml")
# Создание словаря, где ключами являются id рецептов, а значениями - список шагов рецепта
recipes_dict = {id_.text: flatten([step.text for step in recipe.findAll("step")]) for id_, recipe in zip(ab.findAll("id"), ab.findAll('recipe'))}

# вывод первых четырех ключей и значений полученного словаря
print(list(recipes_dict.items())[:4])

````

3.2 По данным файла `steps_sample.xml` сформируйте словарь следующего вида: `кол-во_шагов_в_рецепте: [список_id_рецептов]`

````python
from nltk import flatten

new_recipes_dict = {}
for key in recipes_dict:
    # Подсчет количества шагов в текущем рецепте
    number_of_steps = len(recipes_dict[key])
    # Добавление текущего ключа в список значений словаря new_recipes_dict по ключу number_of_steps
    new_recipes_dict[number_of_steps] = new_recipes_dict.get(number_of_steps, []) + [key]

# вывод первых четырех ключей и значений полученного словаря
print(list(new_recipes_dict.items())[:4])
````

3.3 Получите список рецептов, в этапах выполнения которых есть информация о времени (часы или минуты). Для отбора подходящих рецептов обратите внимание на атрибуты соответствующих тэгов.

````python
import xml.etree.ElementTree as ET

minutes_or_hours_included = []

for recipe in ab.findAll('recipe'):
    # Проходим по всем шагам в каждом рецепте
    for step in recipe.findAll("step"):
        # Преобразуем шаг в строку и создаем XML-дерево
        xmlstr = str(step)
        root = ET.fromstring(xmlstr)
        # Проверяем, есть ли в атрибутах шага 'has_minutes' или 'has_hours'
        if 'has_minutes' in root.attrib or 'has_hours' in root.attrib:
             # Если да, добавляем идентификатор рецепта в список
            minutes_or_hours_included.append(recipe.find("id").text)
            break

print(minutes_or_hours_included[:10])
````
